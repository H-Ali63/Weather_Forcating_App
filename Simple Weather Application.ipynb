```python
import pickle
from tkinter import *
import tkinter as tk
from geopy.geocoders import Nominatim
from tkinter import ttk, messagebox
from timezonefinder import TimezoneFinder
from datetime import *
import requests
from PIL import Image, ImageTk
import pytz
import time
import requests
import json

root = Tk()
root.title("Weather App")
root.geometry("1000x250+300+300")
root.configure(bg='#57adff')
root.resizable(False,False)

from geopy.geocoders import Nominatim
from timezonefinder import TimezoneFinder
from datetime import datetime
import pytz
import requests

def getWeather():
    city = textfield.get()

    geolocator = Nominatim(user_agent="172.16.0.2")
    location = geolocator.geocode(city)

    if location:
        obj = TimezoneFinder()
        result = obj.timezone_at(lng=location.longitude, lat=location.latitude)
        timezone.config(text=result)
        long_lat.config(text=f"{round(location.latitude, 4)}°N,{round(location.longitude, 4)}°E")

        home = pytz.timezone(result)
        local_time = datetime.now(home)
        current_time = local_time.strftime("%H:%M%p")
        clock.config(text=current_time)

        # Weather API
        api_key = "ee1c11ede41b7026ce28c3c9aa1bee1a"
        api_url = f"http://api.openweathermap.org/data/2.5/weather?lat={location.latitude}&lon={location.longitude}&units=metric&exclude=hourly&appid={api_key}"

        
    try:
        json_data = requests.get(api_url).json()

        # Extracting weather information
        if 'main' in json_data:
            temp = json_data['main'].get('temp')
            humidity = json_data['main'].get('humidity')
            pressure = json_data['main'].get('pressure')
        else:
            temp = humidity = pressure = "N/A"

        if 'wind' in json_data and 'speed' in json_data['wind']:
            wind_speed = json_data['wind']['speed']
        else:
            wind_speed = "N/A"

        if 'weather' in json_data and len(json_data['weather']) > 0:
            description = json_data['weather'][0].get('description')
        else:
            description = "N/A"

    except Exception as e:
        print(f"Error: {e}")
        
    t.config(text=(temp,"°C"))
    h.config(text=(humidity,"%"))
    p.config(text=(pressure,"hPa"))
    w.config(text=(wind_speed,"m/s"))
    d.config(text=description)

#### icon
image_icon=PhotoImage(file="C:\\Users\\haida\\Weather App\\Images\\logo.png")
root.iconphoto(False,image_icon)

Round_box=PhotoImage(file="C:\\Users\\haida\\Weather App\\Images\\Rounded Rectangle 1.png")
Label(root,image=Round_box,bg='#57adff').place(x=30,y=100)

# label
label1=Label(root,text="Temperature", font=('Hekvetica',11),fg='White',bg="#203243")
label1.place(x=50,y=110)
        
label2=Label(root,text="Humidity", font=('Hekvetica',11),fg='White',bg="#203243")
label2.place(x=50,y=130)
label3=Label(root,text="Pressure", font=('Hekvetica',11),fg='White',bg="#203243")
label3.place(x=50,y=150)
label4=Label(root,text="Wind Speed", font=('Hekvetica',11),fg='White',bg="#203243")
label4.place(x=50,y=170)
label5=Label(root,text="Description", font=('Hekvetica',11),fg='White',bg="#203243")
label5.place(x=50,y=190)

# search box
search_image=PhotoImage(file="C:\\Users\\haida\\Weather App\\Images\\Rounded Rectangle 3.png")
myimage=Label(image=search_image,bg='#57adff')
myimage.place(x=270, y=120)
weat_image=PhotoImage(file="C:\\Users\\haida\\Weather App\\Images\\Layer 7.png")
weatherimage=Label(root, image=weat_image,bg='#203243')
weatherimage.place(x=290,y=127)

textfield=tk.Entry(root, justify='center', width=15, font=('poppins',25,'bold'),bg="#203243",border=0,fg='white')
textfield.place(x=370,y=130)
textfield.focus()

search_icon=PhotoImage(file="C:\\Users\\haida\\Weather App\\Images\\Layer 6.png")
myimage_icon=Button(image=search_icon, borderwidth=0, cursor="hand2",bg="#203243",command=getWeather)
myimage_icon.place(x=645,y=125)

# Clock (here we will place time)
clock=Label(root,font=('Helvetica',30,'bold'),fg='White',bg="#57adff")
clock.place(x=30,y=20)

# timezone
timezone=Label(root,font=('Helvetica',20,'bold'),fg='White',bg="#57adff")
timezone.place(x=700,y=20)
long_lat=Label(root,font=('Helvetica',20,'bold'),fg='White',bg="#57adff")
long_lat.place(x=700,y=50)

# thpwd
t=Label(root,font=('Helvetica',11),fg='White',bg='#203243')
t.place(x=150,y=110)
h=Label(root,font=('Helvetica',11),fg='White',bg='#203243')
h.place(x=150,y=130)
p=Label(root,font=('Helvetica',11),fg='White',bg='#203243')
p.place(x=150,y=150)
w=Label(root,font=('Helvetica',11),fg='White',bg='#203243')
w.place(x=150,y=170)
d=Label(root,font=('Helvetica',11),fg='White',bg='#203243')
d.place(x=150,y=190)
root.mainloop()
```
